```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
df1['A'].hist()
```

Notice how the hist() method is calling matplotlib library under the hood

```python
df1['A'].plot(kind='hist', bins = 30)
```

```python
#The method we are going to be using will be:
df1['A'].plot.hist()
```

```python
df2.head()
```

```python
df2.plot.area(alpha = 0.4) #area plot where alpha = opacity
```

```python
df2.plot.bar(stacked = True) #Default stack=False
```

When the rows are index values, it just plots them into a bar chart

```python
df1.head()
```

```python
df1.plot.line(y='B',figsize=(12,3),lw=1) #everything from matplotlib is applicable
```

```python
df1.plot.scatter(x='A', y='B', c='C')
```

This is a 3 dimensional graph where color represents the third axis

```python
#we can also do
df1.plot.scatter(x='A', y='B', s = df1['C']*100)
```

```python
df2.plot.box() #box and whiskers
```
