```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
import matplotlib.pyplot as plt
```

# Grids

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris) #Stripped down version of pairplot
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot) #diagonal
g.map_upper(plt.scatter) #upper part of the grid
g.map_lower(sns.kdeplot) #lower part of the grid
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(sns.distplot, 'total_bill')
#we are mapping the characterstics on rows and column and each of the grid contains the a certain type of demographic
#For Example: the first graph belongs to smokers who are getting lunch
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```

# Regression Plots

```python
tips.head()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex') 
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o','v'],
          scatter_kws={'s':50}) 
#scatter_kws calls the matplotlib library on top of which the seaborn library is built and we can customize the plot.
#you generally wont remember these things off the top of your head so use the documentation
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex',
          aspect = 0.6, height = 5) #basically aspect ratio and height of the grid
```

```python
sns.set_style('whitegrid') #you can set what kind of style you want for the 
sns.countplot(x = 'sex', data = tips)
sns.despine() #removing the lines on the top and on the side. Press shift tab to see which ones you can remove
#top andd right are true by default
```

```python
plt.figure(figsize=(12,3)) #pass in a tuple of width and height
sns.countplot(x = 'sex', data = tips)
```

```python
sns.set_context('notebook') #if context is set to poster, it assumes you are going to use for poster so it creates huge text
sns.countplot(x = 'sex', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', palette='coolwarm')
```
