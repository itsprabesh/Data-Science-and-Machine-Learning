```python
import pandas as pd
import numpy as np
from plotly import __version__
print(__version__)
```

```python
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
#Data
df = pd.DataFrame(np.random.randn(100,4), columns = 'A B C D'.split())
df.head()
#We are basically creating a DataFrame of 400 randon numbers from the normal distribution table
```

```python
df2 = pd.DataFrame({'Category':'A B C'.split(), 'Values': [32,43,50]})
```

```python
df2
```

```python
%matplotlib inline
```

```python
df.iplot() #interactive plot
```

```python
df.iplot(kind='scatter', x = 'A', y = 'B', mode = 'markers') 
#by default the iplot method tries and joins all the lines so you have to set the mode to markers to get individual points
```

```python
df2.iplot(kind = 'bar', x = 'Category', y = 'Values') 
#always mention kind = 'something' for what kind of plot you want
```

```python
df
```

```python
#you can call some aggregate function so that iplot is useful
df.sum().iplot(kind='bar') #taking the sum of all categories and plotting it
```

```python
df.iplot(kind='box') #click on index to toggle view
```

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,20,10], 'z':[5,4,3,2,1]})
```

```python
df3
```

```python
df3.iplot(kind = 'surface') #use colorscale = 'something' to change the color. Check plotly documentation for colornames
```

```python
df['A'].iplot(kind = 'hist', bins = 30)
```

```python
df.iplot(kind = 'hist') #click on index to toggle on and off
```

```python
df[['A', 'B']].iplot(kind='spread')
```

```python
df.iplot(kind = 'bubble', x = 'A', y = 'B', size = 'C')
```
