```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

# Loading the data

```python
data = pd.read_csv("simple_linear_regression.csv")
```

The imported data is automatically converted to a DataFrame

```python
data
```

```python
data.describe()
```

# Creating first Regression

Exploring the Data

```python
#Selecting GPA as the dependent variable and SAT as the independent variable, where X-axis: SAT and Y-axis: SAT
y= data["GPA"]
x1 = data["SAT"]
```

```python
#Exploring the data
plt.scatter(x1, y)
plt.xlabel("SAT", fontsize = 20)
plt.ylabel("GPA", fontsize = 20)
plt.show()
```

```python
#Regression
x= sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter (x1,y)
yhat = 0.0017*x1 + 0.2750
fig = plt.plot(x1, yhat, lw =3, c="red", label = "regression line")
plt.xlabel("SAT", fontsize = 20)
plt.ylabel("GPA", fontsize = 20)
plt.show()
```
