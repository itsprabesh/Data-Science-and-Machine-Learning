```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0,5,11)
y= x**2
```

```python
#Functional method of creating a matplotlib
plt.plot(x,y)
#use plt.show if you are using anything other than a jupyter notebook
```

```python
plt.plot(x,y)
plt.xlabel('X Label') 
plt.ylabel('Y Label')
#X and Y labels
```

```python
#Multiplots in same canvas
plt.subplot(1,2,1) #(rows, columns, figure you are working on)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented method
fig = plt.figure() #figure object created
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #pass in a list of dimensions of the axes, where [left, bottom, width, height]
axes.plot(x,y) #plot x and y
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #first axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #second axes with different dimensions
axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x, 'r')
axes2.set_title('Smaller Plot')
```

So, basically we are creating a blank canvas usinf plt.figure() and we are creating a set of axes with different scales and mait it look like one is inside of the other. You can pass in the list of dimensions play around with the numbers to see how it changes. 

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x,y)
for current_ax in axes:
    current_ax.plot(x,y)
#axes is an array of axes objects
```

```python
#RRy means that you can index it so you can access one exes at a time
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title("Figure 1")
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')
axes[1].plot(y,x)
axes[1].set_title("Figure 2")
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')
plt.tight_layout()
```

# Figure Size and DPI

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8,2), dpi = 200) #tuple is the height and width in inches
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout() #prettify
```

```python
#LEGENDS

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = "x squared") #specify a label when you plot the graph
ax.plot(x, x**3, label = "x cubed")

ax.legend(loc = 10) #call the legend function on the set of axes; location codes: 0 through 10
# you can also pass in a tuple specifying the location
```

# Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'green') # the color can be anything, regular color or a hex code for color like FF8C00
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'green', linewidth =3) #LINEWIDTH, you can also type in lw
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'green', lw =3, alpha = 0.5) #alpha is the opacity of the line
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'green', lw =3, linestyle = '--') #you can use -, -- or steps or a lot other; can also use ls = "--"
```

```python
#Markers
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'green', lw =2, ls = '--', marker = "o", markersize = 10,
       markerfacecolor = 'red', markeredgewidth = 3, markeredgecolor = 'green') 
#marking the points of (x,y); can be anything * or o or whatever is allowed
```

# Examples

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
